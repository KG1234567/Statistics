```python
import numpy as np
from scipy import stats
```

```python
# CI

sample_mean = 31
pop_std = 5
n = 36
confidence = 0.95
```

```python
se = pop_std / np.sqrt(n)
```

```python
# Calculate the Interval

lower, upper = stats.norm.interval(confidence, loc=sample_mean, scale=se)
```

```python
print(f"95% Confidence Interval: [{lower:.2f}, {upper:.2f}]")
```

## Test

```python
bulb_data = [980, 960, 1020, 940, 950]
claime_mean = 1000

t_stat, p_val = stats.ttest_1samp(bulb_data, claime_mean)

print(f"T-statistic: {t_stat:.2f}, P-value: {p_val:.4f}")
```
