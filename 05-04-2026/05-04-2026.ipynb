```python
def detect_outliers(data, k=1.5):
    data.sort()
    arr = np.array(data, dtype=float)
    q1 = np.percentile(arr, 25, method='linear')
    q3 = np.percentile(arr, 75, method='linear')
    iqr = q3 - q1
    lower_bound = q1 - k * iqr
    upper_bound = q3 + k * iqr
    print(lower_bound, upper_bound)
    outliers = [x for x in arr if x < lower_bound or x > upper_bound]
    return outliers


student_scores = [10, 12, 45, 34, 20, -9, 94, 33, 35, 40, -3, -4, -21, 55, 66, -78, 61, 62, 68, 70, 44, 48, 53, 90, 98, 178, 782]
```

Method - 2

```python
def detect_outliers(data, k=1.5):
    data.sort()
    arr = np.array(data, dtype=float)
    q1 = np.percentile(arr, 25, method='linear')
    q3 = np.percentile(arr, 75, method='linear')
    iqr = q3 - q1
    lower_bound = q1 - k * iqr
    upper_bound = q3 + k * iqr
    mask = (arr >= lower_bound) & (arr <= upper_bound)   # to remove outliers
    outliers_mask = ~mask  # to get outliers
    return {
        "q1": q1,
        "q3": q3,
        "iqr": iqr,
        "lower_bound": lower_bound,
        "upper_bound": upper_bound,
        "outliers": arr[outliers_mask].tolist(),
        "clean_data": arr[mask].tolist()

    }


student_scores = [10, 12, 45, 34, 20, -9, 94, 33, 35, 40, -3, -4, -21, 55, 66, -78, 61, 62, 68, 70, 44, 48, 53, 90, 98, 178, 782]
```

```python
detect_outliers(student_scores)
```

Kurtosis

```python
data = [10, 20, 30, 40, 97]

mean = np.mean(data)
std = np.std(data)

skew = np.mean((data - mean) ** 3) / std ** 3
kurtosis = np.mean((data - mean) ** 4) / std ** 4
print("Skewness:", skew)
print("Kurtosis:", kurtosis)
```

```python
from scipy.stats import kurtosis

data = [30, 35, 36, 37, 38, 42, 43, 45, 50]

mean = np.mean(data)
std = np.std(data)

skew = np.mean((data - mean) ** 3) / std ** 3

# Pearson Kurtosis (Normal Distribution = 3.0)
kur = np.mean((data - mean) ** 4) / std ** 4

# Excess Kurtosis (Normal Distribution = 0.0)
excess_kurtosis = (np.mean((data - np.mean(data)) ** 4) / (np.std(data) ** 4)) - 3

print("Skewness:", skew)
print("Kurtosis:", kur)
print("Excess Kurtosis:", excess_kurtosis)
print("Kurtosis (scipy):", kurtosis(data))
```
