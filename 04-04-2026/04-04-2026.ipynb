## 1/n vs 1/(n-1)

```python
np.random.seed(88)


population = np.random.normal(loc=50, scale=10, size=100000)

# normal distribution
# loc --> mean --> 50
# scale --> std --> 10
# size --> number of samples --> 100000
```

```python
population[0:10]
```

```python
true_variance = np.var(population)     # 1/n
print("true Population Variance: ", true_variance)
```

```python
n_samples = 1000
sample_size = 5

var_using_n = []
var_using_n_minus_1 = []
var_using_n_minus_2 = []
```

```python
""" sample = np.random.choice(population, size=sample_size, replace=False)
var_using_n = np.var(sample)        # 1/n
var_using_n_minus_1 = np.var(sample, ddof=1)  # 1/(n-1)  
print(var_using_n)
print(var_using_n_minus_1) """
```

```python
for _ in range(n_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    var_using_n.append(np.var(sample))           # 1/n
    var_using_n_minus_1.append(np.var(sample, ddof=1))   # 1/(n-1)  
    var_using_n_minus_2.append(np.var(sample, ddof=2))   # 1/(n-2)

print(var_using_n)
print(len(var_using_n))
print(var_using_n_minus_1)
print(len(var_using_n_minus_1))
print(var_using_n_minus_2)
print(len(var_using_n_minus_2))
```

```python
avg_var_n = np.mean(var_using_n)
avg_var_n1 = np.mean(var_using_n_minus_1)    # 1/(n-1) is an unbiased estimator of population variance
avg_var_n2 = np.mean(var_using_n_minus_2)
```

```python
print("Avergae variance using n  :", avg_var_n)
print("Avergae variance using n-1  :", avg_var_n1)
print("Avergae variance using n-2  :", avg_var_n2)
```
